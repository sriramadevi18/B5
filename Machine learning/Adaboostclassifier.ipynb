```python
# Basic libraries
import numpy as np
import pandas as pd

# Scikit-learn modules
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, classification_report
from sklearn.datasets import load_breast_cancer
```

```python
# Load dataset
data = load_breast_cancer()
X = pd.DataFrame(data.data, columns=data.feature_names)
y = data.target
```

```python
# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
# Base estimator: Decision Tree (stump)
base_estimator = DecisionTreeClassifier(max_depth=1)
```

```python
# Create AdaBoost classifier
adaboost = AdaBoostClassifier(estimator=base_estimator, n_estimators=50, learning_rate=0.5, random_state=42)
```

```python
adb = adaboost.fit(X_train,y_train)
```

```python
adb.score(X_test, y_test)
```

```python
y_pred = adb.predict(X_test)
```

```python
# Evaluate
print("Accuracy (Basic Model):", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))
```

```python
# Parameter grid
param_grid = {
    'n_estimators': [50, 100, 200],
    'learning_rate': [0.01, 0.05, 0.5, 1],
    'algorithm': ['SAMME', 'SAMME.R']
}
```

```python
# AdaBoost with Decision Tree as base estimator
ada = AdaBoostClassifier(estimator=DecisionTreeClassifier(), random_state=42)
```

```python
# GridSearchCV
grid_search = GridSearchCV(estimator=ada,
                           param_grid=param_grid,
                           cv=5,            # 5-fold cross-validation
                           scoring='accuracy',
                           n_jobs=-1,        # Use all CPU cores
                           verbose=2)
```

```python
grid_search.fit(X_train,y_train)
```

```python
# Best parameters
print("Best Parameters:", grid_search.best_params_)
```

```python
print("Best Score on Training Set:", grid_search.best_score_)
```

```python
# Best estimator from GridSearch
best_ada = grid_search.best_estimator_
```

```python
# Predict
y_pred_best = best_ada.predict(X_test)

# Evaluate
print("Accuracy (Tuned Model):", accuracy_score(y_test, y_pred_best))
print("Classification Report:\n", classification_report(y_test, y_pred_best))
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
# Parameter grid
param_grid = {
    'n_estimators': [50, 100, 200,300],
    'learning_rate': [0.01, 0.05, 0.1, 0.5, 1],
    'algorithm': ['SAMME', 'SAMME.R']
}
```

```python
# AdaBoost with Decision Tree as base estimator
ada = AdaBoostClassifier(estimator=DecisionTreeClassifier(), random_state=42)
```

```python
# GridSearchCV
grid_search = GridSearchCV(estimator=ada,
                           param_grid=param_grid,
                           cv=5,            # 5-fold cross-validation
                           scoring='accuracy',
                           n_jobs=-1,        # Use all CPU cores
                           verbose=2)
```

```python
random_cv = RandomizedSearchCV(estimator=ada,
                  param_distributions=param_grid,
                  n_iter=30,
                scoring='accuracy',
                n_jobs=-1,
                   verbose=1 )
```

```python
random_cv.fit(X_train,y_train)
```

```python
random_cv.best_estimator_
```

```python
best_random = random_cv.best_estimator_
```

```python
y_pred_rand = best_random.predict(X_test)
```
